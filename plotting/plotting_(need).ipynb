## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75    #can include a list of errors such that
                                # each error bar length is diff

# Let's start plotting
plt.plot(x, y_1, color='purple', linestyle='dotted', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, #adding error bars
             color = "green", label = "Y3 with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.5)  #opacity of grid lines
plt.legend(loc='upper left')
plt.show()
```

### 1.5 Better with NumPy

```python
# Some data for plotting
import numpy as np
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()

#it is too zoomed in 
```

### 1.7 We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 6, 100) #Return evenly spaced numbers over a specified interval.
#Returns num evenly spaced samples, calculated over the interval [start, stop].
#(start, stop, num of intervals)
x2 = x**2
sin_x = np.sin(x)
plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
plt.style.use('default') #to reset style
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') #horizontal line at y=0
plt.plot(date, anomaly, alpha=.5)                               #requires a starting and ending x value
plt.ylim([-8, 8]) # plot range symmetrical in the y direction
```

```python
plt.style.use('fivethirtyeight') #standard style template

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
```

## Exercise 2 :  A simple mathematical plot

y = e^(-x/10)Sin(x)

```python
empty = []
for x in range(0,51):
    empty.append(x)
x = np.array(empty)
x = np.linspace(0, 50, 500)
y = np.exp(((-x)/10))*np.sin(x)

plt.plot(x, y, color='darkblue')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15, rotation=360)
plt.ylim([-1,1])
plt.title('$\mathregular{{e^{{-x}/10}}}$sin(x)', fontsize =20)

plt.grid(alpha=.25)
#plt.legend(loc='lower right')
#save_file_figure(plt, argv)                  # Hide
plt.show(block=False)

#omt i wna cri
```

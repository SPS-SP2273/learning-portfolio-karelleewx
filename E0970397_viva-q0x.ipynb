```python
img_data_flattened = img_data.flatten() #flatten() converts the 2D numpy array into a 1D array weishenme tho
img_data_flattened
```

```python
plt.hist(img_data_flattened,
         bins=195, label = ['intensities']       # 195 bins to split the data points to every single intensity point                 
         )
plt.xlabel('Intensity of pixel')     
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
